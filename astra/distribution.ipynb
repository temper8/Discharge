### Чтение distribution.dat из файла data/distribution.dat

```python
import os
from collections import namedtuple
def read_distribution(f):
    file_path = os.path.abspath(f)
    file = open(file_path)

    data = { 'X': [] ,'Y': [] }
    

    print(data)

    lines = file.readlines()
    table = []
    for line in lines:
        #print(line)
        table.append(line.split())

    for row in table:
        for index, (p, item) in enumerate(data.items()):
            #print(p, item, index)
            item.append(float(row[index]))

    return data  
```

```python
data = read_distribution('data/distribution.dat')
len(data['X'])
```

```python
x0 = 0
lines_list = []
line = { 'X': [] ,'Y': [] }
for x, y in  zip(data['X'], data['Y']):
    if x0<x:
        if y>0:
            line['X'].append(x)
            line['Y'].append(y)
        x0 = x
    else:
        lines_list.append(line)
        line = { 'X': [x] ,'Y': [y] }
        x0 = x

print(len(lines_list))
len(lines_list[0]['X'])
```

```python
import math
for line in lines_list:
    line['logY'] = [math.log(y) for y in line['Y'] ]
```

```python
%matplotlib widget
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots(1,1, figsize=(15, 10))
for line in lines_list[1350:2200]:
    ax1.plot(line['X'], line['logY']);
#ax1.plot(x, sp_neg['Y']);
#ax2.plot(sp_pos['X'], sp_pos['Y']);
```
